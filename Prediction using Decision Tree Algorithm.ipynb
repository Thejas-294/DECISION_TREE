### DATASET : IRIS dataset available in sklearn

### Step 1 : Importing neccessary libraries/packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
```

### Step 2 : Importing the dataset

```python
data = load_iris()
data
```

### Step 3 : Assigning features to variables

```python
X , Y = data.data , data.target
```

### Step 4 : Spliting the dataset into testing and training data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size = 0.8,random_state = 43)
```

### Step 5 : Training the model

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,Y_train)
```

### Step 6 :Visualizing the tree

```python
tree.plot_tree(clf)
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

### Step 7 : Testing the dataset 

```python
pred = clf.predict(X_test)
pred
```

### Step 8 : Evaluating accuracy

```python
from sklearn.metrics import accuracy_score
model_accuracy = accuracy_score(Y_test,pred)
print("Model Accuracy = ", model_accuracy)
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(Y_test,pred,average= "weighted")
print("f1 score = ", f1)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_test,pred))
```

```python
print(confusion_matrix(Y_test,pred))
```
